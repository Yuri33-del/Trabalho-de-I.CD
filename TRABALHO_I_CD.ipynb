```python
import pandas as pd

nome_do_arquivo = 'produtos.T2.csv'

df = pd.read_csv(nome_do_arquivo, decimal=',')

print(df.columns)

df['PREÇO ATUAL'] = df['PREÇO ATUAL'].astype(str)

df['PREÇO ATUAL'] = df['PREÇO ATUAL'].str.replace(',', '.')
df['PREÇO ATUAL'] = df['PREÇO ATUAL'].str.replace('R$', '')
df['PREÇO ATUAL'] = df['PREÇO ATUAL'].astype(float)

df['PREÇO ATUAL']

media_geral = df['PREÇO ATUAL'].mean()
desvio_padrao = df['PREÇO ATUAL'].std()

print('media_geral:', media_geral)
print('desvio_padrao:', desvio_padrao)

mediana = df['PREÇO ATUAL'].median()
print('mediana:', mediana)

máximo = df['PREÇO ATUAL'].max()
minimo = df['PREÇO ATUAL'].min()

print('máximo:', máximo)
print('minimo:', minimo)

amplitude = máximo - minimo
print('a amplitude é:', amplitude)

grouped = df.groupby('MARCA')['PREÇO ATUAL']
grouped.agg(['mean', 'std'])

result = df.groupby('MARCA')['PREÇO ATUAL'].agg(['mean', 'std'])
result.columns = ['Média', 'Desvio Padrão']

result.plot.bar()
```
